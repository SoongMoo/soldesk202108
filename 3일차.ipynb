```python
'''4. 인자값도 없고 결과값도 없다.'''
def lottoprint():
    lotto = []
    i = 1
    while i <= 45:
        lotto.append(i)
        i += 1

    j = 1
    lottoSize = 45
    while j <= 6:
        lottoSize -= 1
        idx = random.randint(0,lottoSize) # 0~44 까지 index
        num = lotto.pop(idx)  # 7, 8 , 9
        print(str(num)  , end= ',')
        j += 1
```

```python
with open("test1.txt", "w" ) as f:
    for i in range(1, 11):
        f.write("%d 번째입니다\n" % i)

with open("test1.txt", "r" ) as f: ### with를 사용하면 close()를 적지 않아도 됨.
    data = f.read()
print(data)

with open("test1.txt", "a") as f:
    for i in range(11, 21):
        f.write("%d번 째 입니다.\n" % i)

with open("test1.txt", "r") as f:
    data = f.read()
print(data)
```

```python
#### class는 반복적으로 사용해야 하는 변수나 함수가 있는 경우 
### 한번 저장하여 반복적으로 사용하기 위해.

result = 0 # 3, 7
def add(num):
    global result
    result += num
```

```python
add(3)
print(result)
add(4)
print(result)
add(6)
print(result)
```

```python
result1 = 0
def add1(num):
    global result1
    result1 += num
```

```python
add1(3)
print(result1)
add1(4)
print(result1)
add1(6)
print(result1)
```

```python
class ClassAdd: # class란 : 변수와 함수를 묶어 놓은 것
    result = 0
    def add(self, num):
        self.result += num
```

```python
add1 = ClassAdd() # add1 : result, add()
add1.add(2)
print(add1.result)
add1.add(4)
print(add1.result)
add1.add(6)
print(add1.result)
```

```python
add2 = ClassAdd() # add2 : result,add()
add2.add(2)
print(add2.result)
add2.add(4)
print(add2.result)
add2.add(6)
print(add2.result)
```

```python
add3 = ClassAdd() # result ,add
add3.add(2)
print(add3.result)
add3.add(5)
print(add3.result)
add3.add(6)
print(add3.result)
```
